# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df1['region'] = le.fit_transform(df1['region'])
df1['sex'] = le.fit_transform(df1['sex'])
df1.head()
```

```python
#using for loop
for col in df2.columns:
    if df2[col].dtype == np.number:
        continue
    else:
        df2[col] = le.fit_transform(df2[col])
    
```

```python
df2.head()
```

```python
#another way using for loop
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in df3.columns:
    if is_numeric_dtype(df3[col]):
        continue
    df3[col] = le.fit_transform(df3[col])
```

```python
df3.head()
```

# One Hot Encoding

```python
df4.head()
```

```python
dummy_region = pd.get_dummies(df4.region)
dummy_sex = pd.get_dummies(df4.sex)
dummy_smoker = pd.get_dummies(df4.smoker)
dummy_region.head()
```

```python
df4.drop('region',axis= 1,inplace = True)
df4.drop('sex',axis = 1,inplace = True)
df4.drop('smoker',axis = 1,inplace = True)
```

```python
df4 = pd.concat([df4,dummy_region,dummy_sex,dummy_smoker] ,axis = 1)
df4.head()
```

```python
dumy = pd.get_dummies(df6.region,drop_first = True)
```

```python
#using for loop

for col in df7.columns:
    if is_numeric_dtype(df6[col]):
        continue
    new = pd.get_dummies(df6[col],drop_first = True)
    df7.drop(df7[[col]], axis = 1,inplace = True)
    df7 = pd.concat([df7,new],axis = 1)
```
